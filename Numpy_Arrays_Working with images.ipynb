###### When copying an array (eg. a=b) any change in b will be reflected in a and vice-versa

```python
a = np.array([1,2,3])
print(a)
b=a
print(b)
```

```python
b[2] = 100
print(b)
print(a) # Any change in b will be reflected in a
```

###### To prevent any change to be reflected in the original array we can use copy keyword

```python
b = a.copy()
print(b)
```

```python
b[1] = 10
print(b)
print(a)
```

###### Reorganizing arrays

```python
before = np.random.randint(2,10,size = (4,3))
print(before)

after = before.reshape((2,2,3))
print(after)
```

###### Vertically stacking matrices

```python
v1 = np.random.randint(2,10,size = (2,3))
print(v1)

v2 = np.random.randint(11,20,size = (2,3))
print(v2)

print(100*'-')

print(np.vstack([v1,v2]))

np.stack((v1,v2,v1,v1))
```

###### Horizontal stack of matrices

```python
h1 = np.random.randint(1,5,size = (2,4))
print(h1)
h2 = np.random.randint(6,10,size = (2,4))
print(h2)

print(50*'=')

print(np.hstack([h1,h2]))

print(50*'=')

print(np.hstack((h2,h1,h1,h2)))
```

## Working with an image using numpy

```python
from skimage import io
photo = io.imread('NumPy_Image.jpg')
photo
```

```python
photo.shape
```

```python
import matplotlib.pyplot as plt
# printing original image
plt.imshow(photo)
```

```python
# reverse of the image 
plt.imshow(photo[::-1])
```

```python
# Mirror image of the original image
plt.imshow(photo[:,::-1])
```

```python
# extracted a part of the image using indexing
plt.imshow(photo[218:400,100:500])
```

```python
# Half the sized of the image
plt.imshow(photo[::2,::2])
```

```python
sinphoto = np.sin(photo)
sinphoto
```

```python
print(np.sum(photo))
print(np.prod(photo))
print(np.max(photo))
print(np.min(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.argmin(photo))
print(np.argmax(photo))
```

```python
photo_masked = np.where(photo > 100,200,0)
```

```python
plt.imshow(photo_masked)
```

```python
# Transposing the image
plt.imshow(photo[:,:,0].T)
```

```python
sortphoto = np.sort(photo)
sortphoto
```

```python
plt.imshow(sortphoto)
```
